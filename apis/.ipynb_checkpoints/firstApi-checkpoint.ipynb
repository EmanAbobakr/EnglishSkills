```python
# -*- coding: utf-8 -*-
"""GP_GRAMMAR_PROJECT.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1-076XS32RsoJ8cGdTRZgF3oUSBsvSNEt
"""

import os
import spacy
import nltk
import re
from spacy.lang.en.examples import sentences
from spacy.matcher import Matcher
from spacy import displacy
from nltk.corpus import wordnet
nlp = spacy.load("en_core_web_sm")

def Check_Sentence_Errors(sentence):
    if not sentence:
        return "Please write a sentence",[],[]
    label=[] #get every phrase with its name(label) 
    out=[]   #get every label's output to check the errors
    q=0
    
    #preprocessing and get (parsing_tree,pos_list,dep_list)
    tree,pos_list,dep_list,input_list,ps,head_list,root_index=preprocessing(sentence)
    #print(pos_list)
    #print(dep_list)
    #print(tree)
    #print(input_list)
    
    #get every phrase with each output(error or not)
    label,out,start,end,q=sentence_lists(tree,dep_list)
    #print(label)
    #print(out)
    #print(start)
    #print(end)
    
    #get the type of the verb root
    s_type=''
    flabel,findex=find_label(root_index,label,start,end)
    s_type=out[findex][(len(out[findex]))-1]
    return find_errors(label,out,start,end,q,tree,pos_list,dep_list,input_list,head_list,sentence),input_list,ps,s_type

def preprocessing(sentence):
    #convert to lower
    sentence=sentence.lower()
    
    #(remove .) and (replace , with and)
    for i in range(len(sentence)):
        if sentence[i]=='.':
            sentence=insert_string(' ',sentence, i)
        if sentence[i]==',':
            sentence=insert_string(" and ",sentence, i)
            
    #remove spaces from the end
    sentence=sentence.strip()
        
    # Substituting multiple spaces with single space
    sentence = re.sub(r'\s+', ' ', sentence, flags=re.I)
    
    #preprocessing_for_grammar_checking
    pos_list,dep_list,input_list,ps,head_list,root_index=preprocessing_for_grammar_checking(sentence)
    
    #Parsing the sentence
    grammar=Grammar()
    tree=Parsing(pos_list,grammar)
    
    return tree[0],pos_list,dep_list,input_list,ps,head_list,root_index

#tokenization and pos tagging and dependency 
def preprocessing_for_grammar_checking(sentence):
    pos_list=[]
    dep_list=[]
    head_list=[]
    input_list=[]
    #check if there is multi_prepositions
    sentence=multi_prepositions(sentence)
    
    #check if there is multi_question_words
    sentence=multi_question_words(sentence)
    
    #tokenization
    doc = nlp(sentence)
    for token in doc:
        token_list=[]
        
        if token.dep_=="ROOT":
            root_index=token.i
        #get head list
        head_list.append((str(token.head)+" "+str(token.head.i)))
        
        #get pos tag list
        if(len(pos_list)==0 and token.tag_=="VBG"):
            pos_list.append((str(token),"NN"))
        else:
            pos_list.append((str(token),token.tag_))
        
        #get dependency list of each token with indexes
        for child in token.children:
            token_list.append(child.text+" "+str(child.i))
            
        
        #get dependency list of all tokens  [child+" "+child.i]
        dep_list.append(token_list)
    
    
    #get input list
    input_list,ps=Get_Input_List(sentence,pos_list,dep_list)
    return  pos_list,dep_list,input_list,ps,head_list,root_index

def insert_string(new,string, index):
    return string[:index] + new + string[index+1:]

def CAN_BE_INPUT_DATA(pos):
    if not pos=="DET" or pos=="POS" or pos=="WH":
        return 1
    return 0

#check if there is multi_prepositions and put it to the last preposition to make a good pos tag
def multi_prepositions(sentence):
    with open('C:/Users/EmyAbobakr/gehad/multi_prepositions.txt', 'r') as reader:
            for line in reader.readlines():
                lin=line[:len(line)-1].split(" ")
                result = re.search(line[:len(line)-1], sentence)
                if result:
                     sentence=sentence.replace(line[:len(line)-1], lin[len(lin)-1])
    return sentence

#check if there is multi_prepositions and put it to the last preposition to make a good pos tag
def multi_question_words(sentence):
    with open('C:/Users/EmyAbobakr/gehad/multi_question_words.txt', 'r') as reader:
            for line in reader.readlines():
                lin=line[:len(line)-1].split(" ")
                result = re.search(line[:len(line)-1], sentence)
                if result:
                     sentence=sentence.replace(line[:len(line)-1], lin[0])
    return sentence

def Grammar():
    #VP:{<V><ADV>?<V>*(<ADV>*<PP>*<NP>*)*}
    grammar =r""" N:{<UH>|<NN.*>} 
                  V:{<MD|VB.?>}
                  WH:{<W.*>}
                  POS:{<POS>}
                  DET:{<.?DT>}
                  PRON:{<PRP\$?>}
                  ADJ:{<JJ.?>}
                  NUM:{<CD>}
                  ADV:{<RP|EX|W?RB.?>}
                  PREP:{<IN|RP|TO>}
                  CONJ:{<CC>}
                  NEG:{<RB>}
                  PREP:{<IN|RP|TO>}
                  PP:{((<PREP><N>?)|<ADJ>)?<PREP>}
                  ADVV:{<ADV>+(<CONJ>?<ADV>)*}
                  ADJJ:{<ADV>*<ADJ>+(<CONJ>?<ADV>*<ADJ>)*}
                  NP:{<DET>*<PRON>*((<DET><ADVV>*<ADJJ>*)|(<PRON><ADVV>*<ADJJ>*)|(<ADVV>*<ADJJ>*<NUM>)|(<ADVV>*<ADJJ>*<N>))?<NUM>*<ADVV>*<ADJJ>*<N>*(<POS>(<ADVV>*<ADJJ>*<NUM>*<ADVV>*<ADJJ>*<N>+)?)?} 
                  VP:{<V>(<ADVV|ADJJ|NP>*(<PP><NP>)*<WH>?<V>+)?}
                  S:{(<PP>*<NP>)+(<NP|VP|ADJJ|ADVV|PP|PREP|CONJ|NEG|ADV|NUM|ADJ|PRON|DET|POS|WH|V|N>)*}
                  """ 
    return grammar

def Grammar2():
    #VP:{<V><ADV>?<V>*(<ADV>*<PP>*<NP>*)*}
    grammar2 =r""" N:{<UH>|<NN.*>} 
                  V:{<MD|VB.?>}
                  WH:{<W.*>}
                  POS:{<POS>}
                  DET:{<.?DT>}
                  PRON:{<PRP\$?>}
                  ADJ:{<JJ.?>}
                  NUM:{<CD>}
                  ADV:{<RP|EX|W?RB.?>}
                  PREP:{<IN|RP|TO>}
                  CONJ:{<CC|RB|IN|TO>}
                  NEG:{<RB>}
                  PREP:{<IN|RP|TO>}
                  """ 
    return grammar2

def Parsing(pos_list,grammar):
    parser = nltk.RegexpParser(grammar)
    tree = parser.parse(pos_list)
    return tree

def tag_of_first_verb(s):
    if(s[:3]=="inf"):
        return "VB"
    if(s[:3]=="ing"):
        return "VBG"
    if(s[:3]=="md+"):
        return "MD"
    if(s[:3]=="fut"):
        return "MD"
    if(s[:3]=="pre"):
        return "VBZ"
    if(s[:3]=="pas"):
        return "VBD"

def Helping_Verb(verb):
    if (verb=="am" or verb=="'m" or verb=="is" or verb=="'s" or verb=="are" or verb=="'re" or verb=="was" or verb=="were" or verb=="has" or verb=="had" or verb=="have" or verb=="'ve" or verb=="do" or verb=="does" or verb=="did"):
        return 1
    if (verb=="MD"):
        return 1
    return 0

def noun_verb_error(n,vv):
    
    if n=="" or vv=="":
        return "ERROR"+vv
    
    v=vv.split("+")
    
    if (v[0]=="ing"):
            return "ERROR"+vv
        
    if v[0]=="md" or v[0]=="Future":
        return "CORRECT"
        
    if v[0]=="past":
        if n=="i":
            if v[1]!="pb":
                return "CORRECT"
            else:
                return "ERROR"+vv
        if n=="s":
            if v[1]!="pb":
                return "CORRECT"
            else:
                return "ERROR"+vv
        if n=="p":
            if v[1]!="sb":
                return "CORRECT"
            else:
                return "ERROR"+vv
        if n=="d":
            return "CORRECT"
        
    if (v[0]=="present" or v[0]=="inf"):
        if n=="i":
            if (v[0]=="inf" or v[1]=="ib" or v[1]=="pd" or v[1]=="ph" or v[1]=="pv"):
                return "CORRECT"
            else:
                return "ERROR"+vv
        if n=="s":
            if (v[0]=="inf"):
                return "ERROR"+vv
            elif (v[1]=="sb" or v[1]=="sd" or v[1]=="sh" or v[1]=="sv"):
                return "CORRECT"
            else:
                return "ERROR"+vv
        if n=="p":
            if (v[0]=="inf" or v[1]=="pb" or v[1]=="pd" or v[1]=="ph" or v[1]=="pv"):
                return "CORRECT"
            else:
                return "ERROR"+vv
        if n=="d":
            if(v[0]=="inf"):
                return "CORRECT"
            if v[1]=="ib":
                return "ERROR"+vv
            else:
                return "CORRECT"
    if v[0]=="vbn":
         return "ERROR"+vv

def lemma(word):
    doc = nlp(word)
    for token in doc:
        return token.lemma_

def find_if_to_then_VBG(ex,tagged,word):
    sentence=ex.split(" ")
    found=0
    k=0
    st=""
    for i in range(len(sentence)):
        with open('C:/Users/EmyAbobakr/gehad/to_then_VBG.txt', 'r') as reader:
            for line in reader.readlines():
                if(lemma(sentence[i])==line[:len(line)-1]):
                    found=1
        if (sentence[i]=="to" and found==1):
            found=2
        while(tagged[k][1]=="POS"):
            k+=1
        if (tagged[k][1]==word and found==2):
            found=3
            return found
            
        k+=1
    return found

def check_verb_then_VBG(v):
    with open('C:/Users/EmyAbobakr/gehad/verb_then_VBG-VB.txt', 'r') as reader:
            for line in reader.readlines():
                if(lemma(v)==line[:len(line)-1]):
                    return 1
    return 0

def Have_link(word,word_index,dep_list_index,dep_list):
    for child in dep_list[dep_list_index]:
        c=child.split(" ")
        if (c[0]==word and c[1]==str(word_index)):
            return 1
    return 0

def check_the_first_verb(child,typev):
    is_help=0           #helping verp
    is_imperative=0     #imperative sentence
    is_gerund=0         #Gerund verb
    #if the first verb is inf
    if child[0][1]=="VB":
        typev+="inf"
        is_imperative=1
    #end inf
    
    #if the first verb is ing
    elif child[0][1]=="VBG":
        typev+="ing"
        is_imperative=1
    #end ing
    
    #if the first verb is in past
    elif child[0][1]=="VBD":
        typev+="past"
        is_help=1
        if child[0][0]=="did":
             typev+="+dd"
        elif child[0][0]=="was":
            typev+="+sb"
        elif child[0][0]=="were":
            typev+="+pb"       
        elif child[0][0]=="had":
             typev+="+dh"
        else:
            is_help=0
            typev+="+dv"
    #end past
    
    #if the first verb is in present
    elif child[0][1]=="VBZ" or child[0][1]=="VBP":
        typev+="present"
        is_help=1
        if ((child[0][0]=="is" or child[0][0]=="'s") and child[0][1]=="VBZ"):
             typev+="+sb"
        elif (child[0][0]=="does" and child[0][1]=="VBZ"):
            typev+="+sd"
        elif (child[0][0]=="has" and child[0][1]=="VBZ"):
            typev+="+sh"
        elif child[0][1]=="VBZ":
            typev+="+sv"
            is_help=0
        elif ((child[0][0]=="am" or child[0][0]=="'m") and child[0][1]=="VBP"):
            typev+="+ib"
        elif ((child[0][0]=="are" or child[0][0]=="'re") and child[0][1]=="VBP"):
             typev+="+pb"
        elif (child[0][0]=="do" and child[0][1]=="VBP"):
            typev+="+pd"
        elif ((child[0][0]=="have" or child[0][0]=="'ve") and child[0][1]=="VBP"):
            typev+="+ph"
        elif child[0][1]=="VBP":
            typev+="+pv"
            is_help=0
    #end present
    
    #if the first verb is MD
    elif child[0][1]=="MD":  
        is_help=1
        if child[0][0]=="will":
            typev+="Future" 
        else:
            typev+="md"    
    #end MD
    elif child[0][1]=="VBN":
         typev+="vbn"
    
    else:
        return "ERROR "+child[0][0]
    
    return (typev,is_help,is_imperative,is_gerund)

def check_the_second_verbs(first,typev,all):
    typev,is_help,is_imperative,is_gerund=check_the_first_verb(first,"")
    numv=len(all)+1  #number of verbs
    i=0
    #if the first verb is inf
    if typev[:3]=="inf":  
        if first[0][0]=="be": 
            if all[i][0][1]=="VBN":
                if numv==2:
                    return typev+"+p"
                else:
                    return "ERROR "+all[i][0][0]
            else:
                return "ERROR "+all[i][0][0]
            
        elif all[i][0][1]=="VBG":
            if check_verb_then_VBG(first[0][0]):
                if numv==2:
                    return typev+"+a"
                else:
                    if numv==3:
                        if all[i][0][0]=="being":
                            if all[i+1][0][1]=="VBN":
                                if numv==3:
                                    return typev+"+p"
                                else:
                                    return "ERROR "+all[i+2][0][0] 
                            else:
                                return "ERROR "+all[i+1][0][0] 
                        else:
                            return "ERROR "+all[i+1][0][0] 
                    else:
                        return "ERROR "+all[i+2][0][0]
            else:
                return "ERROR "+all[i][0][0] 
        else:
            return "ERROR "+all[i][0][0]
    #end inf -----------------------------------------------------------------
    
    #if the first verb is ing
    
    elif typev[:3]=="ing":
        if first[0][0]=="being":
            if all[i][0][1]=="VBN":
                if numv==2:
                    return typev+"+p"
                else:
                    return "ERROR "+all[i][0][0] 
            else:
                return "ERROR "+all[i][0][0] 
        else:
            return "ERROR "+all[i][0][0] 
    #end ing -----------------------------------------------------------------
    
    #if the first verb is md
    
    elif first[0][1]=="MD":
        if all[i][0][1]=="VB":
            if numv==2:
                return typev+"+a"
            else:
                #check the third verb
                i+=1
                if all[i][0][1]=="VBN":
                    
                    if all[i-1][0][0]=="be":
                        if numv==3:
                            return typev+"+p"
                        else:
                            return "ERROR "+all[i+1][0][0]
                    elif all[i-1][0][0]=="have":
                        if numv==3:
                            return typev+"+a"
                        #check the fourth verb
                        elif a[i][0][0]=="been":
                            i+=1
                            if numv==4:
                                if a[i][0][1]=="VBN":
                                    return typev+"+p"
                                elif a[i][0][1]=="VBG":
                                    return typev+"+a"
                                else:
                                    return "ERROR "+all[i][0][0]
                            elif numv==5:
                                if a[i][0][0]=="being":
                                    if a[i+1][0][1]=="VBN":
                                        return typev+"+p"
                                    else:
                                        return "ERROR "+all[i+1][0][0]
                                else:
                                    return "ERROR "+all[i][0][0]
                            else:
                                return "ERROR "+all[i+2][0][0]
                        else:
                            return "ERROR "+all[i+1][0][0]
                    else:
                        return "ERROR "+all[i-1][0][1] 
                elif all[i][0][1]=="VBG" and typev[:3]=="Fut":
                    if all[i-1][0][0]=="be":
                        if numv==3:
                            return typev+"+a"
                        #check the fourth verb
                        elif all[i][0][0]=="being":
                            if all[i+1][0][1]=="VBN":
                                if numv==4:
                                    return typev+"+p"
                                else:
                                    return "ERROR "+all[i+2][0][0]
                            else:
                                return "ERROR "+all[i+1][0][0]
                        else:
                            return "ERROR "+all[i][0][0]
                    elif all[i-1][0][0]=="have":
                        #check the fourth verb
                        if all[i][0][0]=="being":
                            if numv==4:
                                if all[i+1][0][1]=="VBN":
                                    return typev+"+p"
                                else:
                                    return "ERROR "+all[i+1][0][0]
                            else:
                                return "ERROR "+all[i+2][0][0]     
                        else:
                            return "ERROR "+all[i][0][0]
                    else:
                        return "ERROR "+all[i][0][0]
                else:
                    return "ERROR "+all[i][0][0]
        else:
            return "ERROR "+all[i][0][0] 
    
    #end md ------------------------------------------------------------------
    
    #if the first verb is in present
    elif typev[:3]=="pre":
        if all[i][0][1]=="VB":
            if (first[0][0]=="do" or first[0][0]=="does"):
                if numv==2:
                    return typev+"+a"
                else:
                    return "ERROR "+all[i+1][0][0]
            else:
                return "ERROR "+first[0][0]
            
            
                
        elif all[i][0][1]=="VBG":
            lenn=len(typev)
            if (typev[lenn-1]=="v" and (typev[lenn-2]=="s" or typev[lenn-2]=="p") and check_verb_then_VBG(lemma(first[0][0]))):
                if numv==2:
                    return typev+"+a"
                
                elif numv==3:
                    if all[i][0][0]=="being":
                        if all[i+1][0][1]=="VBN":
                            return typev+"+p"
                        else:
                            return "ERROR "+all[i+1][0][0]
                    else:
                        return "ERROR "+all[i][0][0]
                else:
                    return "ERROR "+all[i][0][0]
                
            elif (first[0][0]=="am" or first[0][0]=="'m" or first[0][0]=="is" or first[0][0]=="'s" or first[0][0]=="are" or first[0][0]=="'re"):
                if numv==2:
                    return typev+"+a"
                else:
                    i+=1
                    if all[i][0][1]=="VBN":
                        if all[i-1][0][0]=="being":
                            if numv==3:
                                return typev+"+p"
                            else:
                                return "ERROR "+all[i][0][0]
                        else:
                            return "ERROR "+all[i-1][0][0]
                    else:
                        return "ERROR "+all[i][0][0]
            else:
                return "ERROR "+all[i][0][0]
            
        elif all[i][0][1]=="VBN": 
            if (first[0][0]=="am" or first[0][0]=="'m" or first[0][0]=="is" or first[0][0]=="'s" or first[0][0]=="are" or first[0][0]=="'re"):
                if numv==2:
                    return typev+"+p"
                else:
                    return "ERROR "+all[i][0][0]
                    
            elif (first[0][0]=="has" or first[0][0]=="have" or first[0][0]=="'ve" or first[0][0]=="'s"):  
                if numv==2:
                    return typev+"+a"
                else:
                    i+=1
                    if all[i][0][1]=="VBN":
                        if all[i-1][0][0]=="been":
                            if numv==3:
                                return typev+"+p"
                            else:
                                return "ERROR "+all[i][0][0]
                        else:
                            "ERROR "+all[i-1][0][0]
                    elif all[i][0][1]=="VBG":
                        if all[i-1][0][0]=="been":
                            if numv==3:
                                return typev+"+a"
                            else:
                                if all[i+1][0][1]=="VBN" and numv==4:
                                    if all[i][0][0]=="being":
                                        return typev+"+p"
                                    else:
                                        return "ERROR "+all[i][0][0]
                                else:
                                    return "ERROR "+all[i+1][0][0]
                        else:
                            return "ERROR "+all[i-1][0][0] 
                    else:
                        return "ERROR "+all[i][0][0]
            else:
                return "ERROR "+all[i][0][0]
        else:
            return "ERROR "+all[i][0][0]  
    #end present
    
    #if the first verb is in past
    elif typev[:3]=="pas":
        if all[i][0][1]=="VB":
            if first[0][0]=="did":
                if numv==2:
                    return typev+"+a"
                else:
                     return "ERROR "+all[i][0][0]
            else:
                 return "ERROR "+first[0][0]
                
        elif all[i][0][1]=="VBG":
            lenn=len(typev)
            if (typev[lenn-1]=="v" and typev[lenn-2]=="d" and check_verb_then_VBG(lemma(first[0][0]))):
                if numv==2:
                    return typev+"+a"

                elif numv==3:
                    if all[i][0][0]=="being":
                        if all[i+1][0][1]=="VBN":
                            return typev+"+p"
                        else:
                            return "ERROR "+all[i+1][0][0]
                    else:
                        return "ERROR "+all[i][0][0]
                else:
                    return "ERROR "+all[i][0][0]
            if (first[0][0]=="was" or first[0][0]=="were"):
                if numv==2:
                    return typev+"+a"
                else:
                    if numv==3:
                        if all[i][0][0]=="being":
                            if all[i+1][0][1]=="VBN":
                                return typev+"+p"
                            else:
                                return "ERROR "+all[i+1][0][0]
                        else:
                            return "ERROR "+all[i][0][0]
                    else:
                        return "ERROR "+all[i][0][0]
            else:
                return "ERROR "+first[0][0]
            
        elif all[i][0][1]=="VBN":
            if (first[0][0]=="was" or first[0][0]=="were"):
                if numv==2:
                    return typev+"+p"
                else:
                    return "ERROR "+all[i][0][0]
                
            elif first[0][0]=="had":
                if numv==2:
                    return typev+"+a"
                else:
                    i+=1
                    if all[i][0][1]=="VBN":
                        if all[i-1][0][0]=="been":
                            if numv==3:                           
                                return typev+"+p"
                            else:
                                return "ERROR "+all[i][0][0]
                        else:
                            return "ERROR "+all[i-1][0][0]
                    elif all[i][0][1]=="VBG":
                        if all[i-1][0][0]=="been":
                            if numv==3:                           
                                return typev+"+a"
                            else:
                                if (all[i][0][0]=="being" and numv==4):
                                    if all[i+1][0][1]=="VBN":
                                        return typev+"+p"
                                    else:
                                        return "ERROR "+all[i+1][0][0]
                                else:
                                    return "ERROR "+all[i][0][0]     
                        else:
                            return "ERROR "+all[i-1][0][0]
                    else:
                        return "ERROR "+all[i-1][1]
            else:
                return "ERROR "+all[i][0][0]          
        else:
             return "ERROR "+all[i][0][0]   
    #end past
    return "ERROR "+all[i][0][0]

#VP:{<V>(<ADVV|ADJJ|NP>*(<WH>?<PP><NP>)*<WH>?<V>+)?}
#verb=[label,verb,pos]

def CHECK_VP_ERRORS(all,index,dep_list):
    # initializations
    typev=""            #the output of the function(the verb tense)
    l=len(all)          #the length of the verp_phrase
    c=0                 #counter of labels
    i=0                 #counter of all words
    is_help=0           #helping verp
    is_imperative=0     #imperative sentence
    is_gerund=0         #Gerund verb
    have_np=0
    #work on the first verb---------------------------------------------------- 
    typev,is_help,is_imperative,is_gerund=check_the_first_verb(all[i],"")
     
    
    #***check if there is only one verp
    if l==1:
        typev+="+a"
        return  typev,1,is_help,is_imperative,is_gerund,have_np
    
    #if i not = l
    i+=1
    c+=1
    #check if there is a NP|ADVV|ADJJ|PP-----------------------------------
    
    #ignore adj and adv
    while (c<l and (all[c].label()=="ADVV" or all[c].label()=="ADJJ")):
        i+=len(all[c])
        c+=1
    while (c<l and ((all[c].label()=="NP") or (c+1<l and all[c].label()=="PP" and all[c+1].label()=="NP"))):
        if all[c].label()=="PP":
            fun=CHECK_NP_ERRORS(all[c+1],index+i,dep_list)
            if fun[0][0]=="E":
                return fun
            
            else:
                if is_help==1:
                    NPfun=noun_verb_error(fun[0],typev)
                    if NPfun[0]=="E":
                        return "ERROR "+all[i][0][0]
                    
            c+=2
            i+=fun[1]+1
        else:
            have_np=1
            fun=CHECK_NP_ERRORS(all[c],index+i,dep_list)
            if fun[0][0]=="E":
                return fun
            
            else:
                if is_help==1:
                    NPfun=noun_verb_error(fun[0],typev)
                    if NPfun[0]=="E":
                        return "ERROR "+all[i][0][0]
            c+=1
            i+=fun[1]
    #check if there is a WH--------------------------------------------------
    while (c<l and all[c].label()=="WH"):
        i+=1
        c+=1
    
    #check the last verbs----------------------------------------------------
    ii=index+i+(l-c-1)    #index of the last word
    root=c+(l-c-1)        #counter for the last label
    i+=(l-c)              #counter for all words to end
    ic=ii                 #counter for the last word
    while root>=c:
        ii-=1
        ch=root-1        #the index of the current root
        sizev=1
        while (ch>=c and Have_link(all[ch][0][0],ii,ii+1,dep_list)==1):
            ii-=1
            ch-=1
            ic-=1
            sizev+=1
    
        #me=ch+1 index=ii+1 indexofroot=ic lastlabel=root
        
        if (Have_link(all[0][0][0],index,ii+1,dep_list)==1 or (Have_link(all[0][0][0],index,ic,dep_list)==1)):
            fs=check_the_second_verbs(all[0],typev,all[ch+1:root+1])
            if(fs[0]=="E"):
                return fs
            else:
                typev=fs
        
        else:
            if  Have_link(all[ch+1][0][0],ii+1,ii,dep_list)==1:
                if not(check_verb_then_VBG(lemma(all[ch][0][0])) and all[ch+1][0][1]=="VBG"):
                     return "ERROR "+all[root][0][0]
                
            typet=check_the_first_verb(all[ch+1],"")
            
            if sizev>1:
                fs=check_the_second_verbs(all[ch+1],typet[0],all[ch+2:root+1])
                if(fs[0]=="E"):
                    return fs
                else:
                    if have_np:
                        NVfun=noun_verb_error(fun[0],typet[0])
                        if NVfun[0]=="E":
                            return "ERROR "+all[ch+1][0][0]
                            
            else:
                if have_np:
                    NVfun=noun_verb_error(fun[0],typet[0])
                    if NVfun[0]=="E":
                        return "ERROR "+all[ch+1][0][0]
        
        
        root=ch 
        ic-=sizev
    return typev,i,is_help,is_imperative,is_gerund,have_np

def type_of_num(num,count,j):
    i=str(j)
    if count==1:
        if num=="1" or num=="one":
            return "s"+i
        else:
            return "p"+i
    else:
        return "p"+i

def type_of_det(det,j):
    i=str(j)
    # (s) singular / (p) plural / (i) can be alone / (n) can not be alone / (d) singular or plural
    
    
    if det=="each" or det=="this" or det=="that":
        return "si"+i
    
    elif det=="a" or det=="an" or det=="every" or det=="another": 
        return "sn"+i
    
    elif det=="these" or det=="those" or det=="all" or det=="some" or det=="others":
        return "pi"+i
    
    elif det=="other":
        return "pn"+i
    
    elif det=="the":
        return "dn"+i
    
    else:
        return "di"+i

def type_of_pron(pron,j):
    i=str(j)
    # 1
    if pron=="me": 
        return "1i"+i
    elif pron=="him" or pron=="her" or pron=="it": 
        return "1s"+i
    elif pron=="us" or pron=="you" or pron=="them":
        return "1p"+i
    
    # 2
    if pron=="mine":
        return "2i"+i
    elif pron=="his" or pron=="hers" or pron=="its":
        return "2s"+i
    elif pron=="ours" or pron=="yours" or pron=="theirs":
        return "2p"+i
    
    # 3
    if pron=="my":
        return "3i"+i
    elif pron=="his" or pron=="her" or pron=="its":
        return "3s"+i
    elif pron=="our" or pron=="your" or pron=="their":
        return "3p"+i
    
    # 4
    if pron=="i":
        return "4i"+i
    elif pron=="he" or pron=="she" or pron=="it":
        return "4s"+i
    elif pron=="we" or pron=="you" or pron=="they":
        return "4p"+i
    
    # 5
    if pron=="myself":
        return "5i"+i
    elif pron=="himself" or pron=="herself" or pron=="itself":
        return "5s"+i
    elif pron=="ourselves" or pron=="yourself" or pron=="yourselves" or pron=="themselves":
        return "5p"+i

def correct_pron(a,b):
    if a=="i":
        if b=="myself":
            return 1
        else:
            return 0
    elif a=="he":
        if b=="himself":
            return 1
        else:
            return 0
    elif a=="she":
        if b=="herself":
            return 1
        else:
            return 0
    elif a=="it":
        if b=="itself":
            return 1
        else:
            return 0
    elif a=="we":
        if b=="ourselves":
            return 1
        else:
            return 0
    elif a=="you":
        if b=="yourself" or b=="yourselves":
            return 1
        else:
            return 0
    elif a=="they":
        if b=="themselves":
            return 1
        else:
            return 0
    else:
        return 0

def CHECK_NP_ERRORS(all,index,dep_list):
    l=len(all)   #the length of the list     
    typeSP=""    #the output
    start=0      #start is the index in NP list
    det=""
    pron=""
    adv=""
    adj=""
    num=""
    noun=""
    
    
    
    #check if NP starts with DET
    
    count=0
    while (start<l and all[start].label()=="DET"):
        count+=1
        #check if there is repetition
        if count>1:
            if all[start].label()==all[start-1].label():
                return "ERROR "+all[start][0][0]
            elif not (re.match(r"the",all[start-1][0][0]) and re.match(r"(all)|(every)|(each)",all[start][0][0])):
                return "ERROR "+all[start][0][0]
        #get the det type
        det = type_of_det(all[start][0][0],start)
        start+=1
    
    #end det
  
    #check if there is only determiner in NP
    if start==l:
        if (det!="" and det[1]=="i"):
            return det[0],l
        else:
            return "ERROR "+all[start-1][0][0]
    elif all[start-1][0][0]=="a":
        if (re.match(r"(a|e|i|o|u)",all[start][0][0][0])):
            return "ERROR "+all[start-1][0][0]
    elif all[start-1][0][0]=="an":
        if not (re.match(r"(a|e|i|o|u)",all[start][0][0][0])):
            return "ERROR "+all[start-1][0][0]
        
    
    #end checking 
    #------------------------------------------------------------------------------
    
    #check if NP starts with PRON
    
    count=0
    while (start<l and all[start].label()=="PRON"):
        count+=1
        #check if there is repetition
        if count>1:
            s1=type_of_pron(all[start-1][0][0],start-1)
            s2=type_of_pron(all[start][0][0],start)
            if all[start][0][0]==all[start-1][0][0]:
                return "ERROR "+all[start][0][0]
            if not ((correct_pron(all[start-1][0][0],all[start][0][0])==1) or (s1[0]=="1" and s2[0]=="5")):
                return "ERROR "+all[start][0][0]
        #get the pron type
        pron = type_of_pron(all[start][0][0],start)
        start+=1
        
    #end PRON
    
    #check if there is only Pronoun in NP
    if start==l:
        if (det!="" and pron!=""):
            return "ERROR "+all[start-1][0][0]
        elif (pron!="" and pron[0]!="3"):
            return pron[1],l
        else:
            return "ERROR "+all[start-1][0][0]
    
    #end checking 
    #------------------------------------------------------------------------------
    
    #check if NP starts with ADV
    
    count=0
    while (start<l and all[start].label()=="ADVV"):
        #if there is repetition then count will be more then 1
        count+=1
        if count>1:
            if all[start][0][0]==all[start-1][0][0]:
                return "ERROR "+all[start][0][0]
        start+=1
        adv="adv"
        
    #end ADV
    
    #check if there is only adverb or not in NP
    if start==l:
        if (det!="" and adv!=""):
            return det[0],l
        elif (pron!="" and adv!=""):
            return pron[1],l
        else:
            return adv,l
    
    #end checking 
    #------------------------------------------------------------------------------
    
    #check if NP starts with ADJ
    
    count=0
    while (start<l and all[start].label()=="ADJJ"):
        #if there is repetition then count will be more then 1
        count+=1
        if count>1:
            if all[start][0][0]==all[start-1][0][0]:
                return "ERROR "+all[start][0][0]
        start+=1
        adj="adj"
        
    #end ADJ
    
    #check if there is only adj or not in NP
    if start==l:
        if (det!="" and adj!=""):
            return det[0],l
        elif (pron!="" and adj!=""):
            return pron[1],l
        else:
            return adj,l
    
    #end checking 
    #------------------------------------------------------------------------------
    
    #check if NP starts with NUM
    
    count=0
    n="" #the all number in string
    numtype=""  #number type
    while (start<l and all[start].label()=="NUM"):
        count+=1
        #check if there is repetition
        if count>1:
            if all[start][0][0]==all[start-1][0][0]:
                return "ERROR "+all[start][0][0]
        n+=all[start][0][0]
        start+=1
        
    #end NUM
    
    #if there is num
    if n!="":
        
        #get the num type
        numtype=type_of_num(n,count,start)
        
        #check if there is det
        if (det!="" and numtype!=""):
            if (numtype[0]=="s" and (det[0]=="s" or det[0]=="d")):
                num=numtype
            elif (numtype[0]=="p" and (det[0]=="p" or det[0]=="d")):
                num=numtype
            else:
                return "ERROR "+all[start-count][0][0]
        
        
        #check if there is pron
        elif (pron!="" and numtype!=""):
            if (pron[0]=="2" or pron[0]=="4"):
                if (all[ int(pron[2]) ][0][0]=="its" or all[ int(pron[2]) ][0][0]=="it" or all[ int(pron[2]) ][0][0]=="you" or all[ int(pron[2]) ][0][0]=="his"):
                    num=numtype
                else:
                    return "ERROR "+all[start-count][0][0]
                    
            else:
                num=numtype
        
        else:
            num=numtype
            
    #check if there is only NUM or not in NP
    if (start==l and n!=""):
        return num[0],l
    
    #end checking 
    #------------------------------------------------------------------------------
    
    
    #ignore any adj or adv
    count=0
    found=0
    while (start<l and (all[start].label()=="ADJJ" or all[start].label()=="ADVV")):
        #if there is repetition then count will be more then 1
        count+=1
        if count>1:
            if all[start][0][0]==all[start-1][0][0]:
                return "ERROR "+all[start][0][0]
        start+=1
        found=1
    
    #check if ends with adj or adv
    if (start==l and found==1):
        return "ERROR "+all[start-1][0][0]
    
    
    
    
    #check if NP starts with N
    
    count=0
    while (start<l and all[start].label()=="N"):
        #if there is repetition then count will be more then 1
        count+=1
        if count>1:
            if all[start][0][0]==all[start-1][0][0]:
                return "ERROR "+all[start][0][0]
        noun=[all[start].label(),all[start][0][0],all[start][0][1]] #+++++++++++++++++++++++++++++++++++++++++++++++++
        start+=1
        
    #end N
    
    #check if there is only N or not in NP
    if start<l:
        #ignore pos
        start+=1
        
        if start==l:
            noun=["N","d","d"]
        else:
            #check if NPos starts with ADV
            count=0
            advpos=""
            while (start<l and all[start].label()=="ADVV"):       
            #if there is repetition then count will be more then 1
                count+=1
                if count>1:
                    if all[start][0][0]==all[start-1][0][0]:
                        return "ERROR "+all[start][0][0]
                start+=1
                advpos="adv"
            #end ADVpos
    
            #check if there is only adverb or not in NPos
            if start==l:
                return "ERROR "+all[start-1][0][0]
    
            #end checking  advpos
        
            #check if NPos starts with ADJ
            count=0
            adjpos=""
            while (start<l and all[start].label()=="ADJJ"):       
            #if there is repetition then count will be more then 1
                count+=1
                if count>1:
                    if all[start][0][0]==all[start-1][0][0]:
                        return "ERROR "+all[start][0][0]
                start+=1
                adjpos="adj"
            #end ADJpos
    
            #check if there is only adj or not in NPos
            if start==l:
                return "ERROR "+all[start-1][0][0]
    
            #end checking  adjpos
            
            
            
            #check if NPos starts with NUM
    
            count=0
            n1="" #the all number in string
            numtype1=""  #number type
            while (start<l and all[start].label()=="NUM"):
                count+=1
                #check if there is repetition
                if count>1:
                    if all[start][0][0]==all[start-1][0][0]:
                        return "ERROR "+all[start][0][0]
                n1+=all[start][0][0]
                start+=1
        
            #end NUM
    
            #if there is num
            if n1!="":
        
                #get the num type
                numtype1=type_of_num(n1,count,start)
                num=numtype1
            #check if there is only NUM or not in NP
            if (start==l and n!=""):
                return numtype1[0],l
    
            #end checking num
        
            #check if NPos starts with ADJ or adv
            count=0
            while (start<l and (all[start].label()=="ADJJ" or all[start].label()=="ADVV")):       
            #if there is repetition then count will be more then 1
                count+=1
                if count>1:
                    if all[start][0][0]==all[start-1][0][0]:
                        return "ERROR "+all[start][0][0]
                start+=1
            #end
    
            #check if there is only adj or adv in NPos
            if start==l:
                return "ERROR "+all[start-1][0][0]
    
            #end checking  adj adv
            
            
            #check if NPos ends with N
    
            count=0
            while (start<l and all[start].label()=="N"):
            #if there is repetition then count will be more then 1
                count+=1
                if count>1:
                    if all[start][0][0]==all[start-1][0][0]:
                        return "ERROR "+all[start][0][0]
                noun=[all[start].label(),all[start][0][0],all[start][0][1]]  #+++++++++++++++++++++++++++++++++++++++++  
                start+=1
        
            #end Npos
            
    #now we have the final noun
    typen="d"
    #check noun type
    if noun[2]=="d":
        typen="d"
    elif noun[2]=="NNS":
        typen="p"
    else:
        typen="s"
    #end checking noun type 
    
    #check if there was a num
    if (typen!="" and num!=""):
        if typen=="d":
            return num[0],l
        elif (num[0]=="s" and typen=="s"):
            return typen,l
        elif (num[0]=="p" and typen=="p"):
            return typen,l
        else:
            return "ERROR "+noun[1]
        
    #end check
    
    #check if there was a det
    if (typen!="" and det!=""):
        if typen=="d":
            return det[0],l
        elif ((det[0]=="s" or det[0]=="d")and typen=="s"):
            return typen,l
        elif ((det[0]=="p" or det[0]=="d") and typen=="p"):
            return typen,l
        else:
            return "ERROR "+noun[1]
        
    #end check
    
    #check if there was a pron
    if (typen!="" and pron!=""):
        if (all[int(pron[2])][0][0]=="his" or all[int(pron[2])][0][0]=="her" or all[int(pron[2])][0][0]=="its" or pron[0]=="3"):
            return typen,l
        else:
            return "ERROR "+all[int(pron[2])][0][0]

    #end check
    
    

    #end checking N
    #------------------------------------------------------------------------------           
    return typen,l

def find_label(index,label,start,end):
    for i in range(len(label)):
        if index>=start[i] and index<=end[i]:
            return label[i],i

def sentence_lists(tree,dep_list):
    label=[]
    out=[]
    start=[]
    end=[]
    i=0
    q=0 
    for child in tree:
            
        if child.label()=="WH":
            label.append("WH")
            #check if the sentence is question
            if i==0:
                q=1
            out.append("WH") 
            start.append(i)
            end.append(i+len(child)-1)
            i+=len(child)
            
        elif child.label()=="VP":
            #typev,i,is_help,is_imperative,is_gerund,have_np
            label.append("VP")
            f=CHECK_VP_ERRORS(child,i,dep_list)
            if f[0][0]=="E":
                out.append(f)
                start.append(i)
                i+=len(child)
                end.append(i-1)
            else:  
                out.append(f[0])
                start.append(i)
                i+=f[1]
                end.append(i-1)
            
            
        elif child.label()=="NP":
            f=CHECK_NP_ERRORS(child,i,dep_list)
            if f[0][0]=="E":
                label.append("NP")
                out.append(f)
                start.append(i)
                end.append(i+len(child)-1)
                i+=len(child)
                
            elif (f[0]=="adj" or f[0]=="adv"):
                label.append(child[0].label())
                out.append(child[0][0][0][0])
                start.append(i)
                end.append(i+len(child[0])-1)
                i+=len(child[0])
            
            else:  
                label.append("NP")
                out.append(f[0])
                start.append(i)
                i+=f[1]
                end.append(i-1)
            
        else:
            label.append(child.label())
            out.append(child[0][0])
            start.append(i)
            end.append(i+len(child)-1)
            i+=len(child)
            
               
    return label,out,start,end,q

def find_errors(label,out,start,end,q,tree,pos_list,dep_list,input_list,head_list,sentence):
    senn=sentence.split(" ")
    result="CORRECT"
    l1=""
    k=-1
    if (label[0]!="NP" and label[0]!="PP"):
        return "please enter a Declarative Sentence"
    for l2 in label:
        k+=1
        if out[k][0]=="E":
            return out[k]
        
        elif l2=="VP":
            if l1=="NP":
                if out[k-1][0]=="E":
                    return out[indx]
                f=noun_verb_error(out[k-1],out[k])
                if f[0]=="E":
                    return "ERROR "+senn[start[k]]
            
            
            if (label[k-2]=="VP" and l1=="CONJ"):
                    
                if out[k-2][0]=="E":
                    return out[k-2]      
                else:
                    if not((out[k]==out[k-2]) or (pos_list[end[k-2]][1]==pos_list[end[k]][1])):
                        return "ERROR "+senn[start[k]] 
            
            
            
            for d in dep_list[end[k]]:
                
                if d=="":
                    continue
                dep=d.split(" ")
                if (start[k]<=int(dep[1])):
                    continue
                lab,indx=find_label(int(dep[1]),label,start,end)
                #check if there is noun_verb_error
                if lab=="NP":
                    if out[indx][0]=="E":
                        return out[indx]
                    f=noun_verb_error(out[indx],out[k])
                    if f[0]=="E":
                        return "ERROR "+senn[start[k]] 
                if lab=="VP" and l1=="CONJ":
                    
                    if out[indx][0]=="E":
                        return out[indx]
                    
                    else:
                        if not((out[k]==out[k-2]) or (pos_list[end[k-2]][1]==pos_list[end[k]][1])):
                            return "ERROR "+senn[start[k]] 
                
            if head_list[end[k]]!="":
                head=head_list[end[k]].split(" ")
                lab,indx=find_label(int(head[1]),label,start,end)
                if lab=="VP" and l1=="CONJ":
                    if out[indx][0]=="E":
                        return out[indx]
                        
                    else:
                        if not((out[k]==out[k-2]) or (pos_list[end[k-2]][1]==pos_list[end[k]][1])):
                            return "ERROR "+senn[start[k]] 
                        
        
        elif ((k>0) and (l1=="PP" and l2=="VP")):
            
            outsplit1=out[k-1].split("_")
            outsplit2=out[k].split("_")
            outsplit3=out[k].split("+")
            
            if out[k-1]!="to":
                if out[k][:3]!="ing":
                    return "ERROR "+senn[start[k-1]]
            else:
                if(k<len(tree)):
                    
                    if(find_if_to_then_VBG(sentence,pos_list,tag_of_first_verb(out[k]))==3):
                        
                        if out[k][:3]!="ing":
                            return "ERROR"+senn[start[k]]
                    else:
                        if out[k][:3]!="inf":
                            return "ERROR to"
                else:
                    return "ERROR "+senn[start[k]]
                
        elif ((k>0) and (l1=="PP" and l2=="NP")):
            if (len(tree[k])==1 and tree[k][0].label()=="PRON"):
                pron = type_of_pron(tree[k][0][0],0)
                if not(pron[0]=="1" or pron[0]=="5"):
                    return "ERROR "+tree[k][0][0]
        
        #conj
        elif ((k>0) and (l2=="CONJ") and (k==len(label)-1)):
            if (k==len(label)-1):
                return "ERROR "+tree[k][0][0]
            
            
        elif ((k>1) and (l1=="CONJ")):
            similar=0
            if label[k]==label[k-2]:
                if label[k]=="NP":
                    for d in dep_list[end[k]]:
                        if d=="":
                            continue
                        dep=d.split(" ")
                        lab,indx=find_label(int(dep[1]),label,start,end)
                        #check if there is noun_verb_error
                        if (indx>=start[k-2] and indx<=end[k-2]):
                            similar=1
                            break
                        
                    for d in dep_list[end[k-2]]:
                        if d=="":
                            continue
                        dep=d.split(" ")
                        lab,indx=find_label(int(dep[1]),label,start,end)
                        #check if there is noun_verb_error
                        if (indx>=start[k] and indx<=end[k]):
                            similar=1
                            break
                        
                        
                    
                    
                    if similar==1:
                        if (out[k-1]=="or"):
                            if (out[k]!="s" or out[k-2]!="s"):
                                out[k]="p"
                        elif (out[k-1]=="and"):
                            out[k]="p"
                        
                        start[k]=start[k-2]
                        label[k-1]=label[k]
                        out[k-1]=out[k]
                        start[k-1]=start[k]
                        end[k-1]=end[k]
                        
                        label[k-2]=label[k]
                        out[k-2]=out[k]
                        start[k-2]=start[k]
                        end[k-2]=end[k]
                    
                
                    
                    
                
        #make l1 = l2
        l1=l2
    return result

def Get_Input_List(sentence,pos_list,dep_list):
    conj=0
    Input_list=[]
    Input_list_index=[]
    ps=[]
    v_v=[]
    sen=sentence.split(" ")
    grammar=Grammar2()
    t_list=Parsing(pos_list,grammar)
    #any adj should be the two 
    for c in range(len(dep_list)):
        for ch in dep_list[c]:
            child=ch.split(" ")
            if(t_list[c].label()=="V" and (t_list[int(child[1])].label()=="N" or t_list[int(child[1])].label()=="PRON" or t_list[int(child[1])].label()=="NUM")):
                if (c>int(child[1])):
                    Input_list.append([child[0],sen[c]])
                    ps.append([t_list[int(child[1])].label(),t_list[c].label()])
                    Input_list_index.append([child[0],int(child[1]),sen[c],c])
                    
            elif(t_list[int(child[1])].label()=="V" and (t_list[c].label()=="N" or t_list[c].label()=="PRON" or t_list[c].label()=="NUM")):
                if (c<int(child[1])):
                    Input_list.append([child[0],sen[c]])
                    ps.append([t_list[int(child[1])].label(),t_list[c].label()])
                    Input_list_index.append([child[0],int(child[1]),sen[c],c])
            
            
            elif(t_list[c].label()=="V" and t_list[int(child[1])].label()=="V"):
                if (c>int(child[1])):
                    if(check_verb_then_VBG(lemma(child[0]))):
                        Input_list.append([child[0],sen[c]])
                        ps.append([t_list[int(child[1])].label(),t_list[c].label()])
                        Input_list_index.append([child[0],int(child[1]),sen[c],c])
                        v_v.append([child[0],sen[c],len(Input_list)-1])
                
                 
                elif (c<int(child[1])):
                    if(check_verb_then_VBG(lemma(sen[c]))):
                        Input_list.append([sen[c],child[0]])
                        ps.append([t_list[c].label(),t_list[int(child[1])].label()])
                        Input_list_index.append([sen[c],c,child[0],int(child[1])])
                        v_v.append([sen[c],child[0],len(Input_list)-1])
                    
        
        
        for ch in dep_list[c]:
            child=ch.split(" ")
            if(t_list[c].label()=="V" and (t_list[int(child[1])].label()=="N" or t_list[int(child[1])].label()=="PRON" or t_list[int(child[1])].label()=="NUM" or t_list[int(child[1])].label()=="ADJ")):
                if (c<int(child[1])):
                    Input_list.append([sen[c],child[0]])
                    ps.append([t_list[c].label(),t_list[int(child[1])].label()])
                    Input_list_index.append([sen[c],c,child[0],int(child[1])])
            elif(t_list[int(child[1])].label()=="V" and (t_list[c].label()=="N" or t_list[c].label()=="PRON" or t_list[c].label()=="NUM" or t_list[c].label()=="ADJ")):
                if (c>int(child[1])):
                    Input_list.append([sen[c],child[0]])
                    ps.append([t_list[c].label(),t_list[int(child[1])].label()])
                    Input_list_index.append([sen[c],c,child[0],int(child[1])])
            elif((t_list[int(child[1])].label()=="N" or t_list[int(child[1])].label()=="PRON" or t_list[int(child[1])].label()=="NUM" or t_list[int(child[1])].label()=="ADJ") and (t_list[c].label()=="N" or t_list[c].label()=="PRON" or t_list[c].label()=="NUM" or t_list[c].label()=="ADJ") and not (t_list[int(child[1])].label()=="ADJ" and t_list[c].label()=="ADJ")):
                if (c<int(child[1])):
                    Input_list.append([sen[c],child[0]])
                    ps.append([t_list[c].label(),t_list[int(child[1])].label()])
                    Input_list_index.append([sen[c],c,child[0],int(child[1])])
                else:
                    Input_list.append([child[0],sen[c]])
                    ps.append([t_list[int(child[1])].label(),t_list[c].label()])
                    Input_list_index.append([child[0],int(child[1]),sen[c],c])
        
    
    
    
    conj=0
    for c in range(len(sen)):
        if t_list[c].label()=="CONJ":    
            conj=1
            continue
        
        
        if conj==1:
            conj=0
            cont=0
            if ((t_list[c].label()=="N" and t_list[c-2].label()=="N") or (t_list[c].label()=="V" and t_list[c-2].label()=="V") or (t_list[c].label()=="ADJ" and t_list[c-2].label()=="ADJ")):
                for index in range(len(Input_list_index)):
                    if cont==1:
                        cont=0
                        continue
                    if Input_list_index[index][1]==c-2:
                        Input_list_index.insert(index+1,[sen[c],c,Input_list_index[index][2],Input_list_index[index][3]])
                        Input_list.insert(index+1,[sen[c],Input_list_index[index][2]])
                        ps.insert(index+1,[t_list[c].label(),t_list[Input_list_index[index][3]].label()])
                        
                        cont=1
                    elif Input_list_index[index][3]==c-2:
                        Input_list_index.insert(index+1,[Input_list_index[index][0],Input_list_index[index][1],sen[c],c])
                        Input_list.insert(index+1,[Input_list_index[index][0],sen[c]])
                        ps.insert(index+1,[t_list[Input_list_index[index][1]].label(),t_list[c].label()])
                        cont=1
    
    
    for vv in range(len(v_v)):
        for index in range(len(Input_list)):
            if index!=v_v[vv][2]:
                if Input_list[index][0]==v_v[vv][0]:
                    Input_list[index][0]=v_v[vv][1]
                    
                if Input_list[index][1]==v_v[vv][0]:
                    Input_list[index][1]=v_v[vv][1]
    
    for index in range(len(ps)):
        if ps[index][0]=="V":
            Input_list[index][0]=lemma(Input_list[index][0])
        if ps[index][1]=="V":
            Input_list[index][1]=lemma(Input_list[index][1])
    return Input_list,ps

def file_data(f2):
    #find unique words in relations file
    s_set=[]
    with open(f2, 'r') as reader:
            for line in reader.readlines():
                lin=line[:len(line)-1]
                s_set+=re.findall("[a-zA-Z]+", lin)
    
            s_set=set(s_set)
            s_list=list(s_set)
            
            
            #fill the matrix with data
            data = [["0" for i in range(len(s_list))] for j in range(len(s_list))]
    with open(f2, 'r') as reader:
            for line in reader.readlines():
                lin=line[:len(line)-1].split('-')
                nodes=lin[2].split("/")
                for nod in nodes:
                    data[s_list.index(lin[0])][s_list.index(nod)]=lin[1]+"=a"
                    data[s_list.index(nod)][s_list.index(lin[0])]=lin[1]+"=p"
    
    return data,s_list

def similar(w1,w2):
    syn1 = wordnet.synsets(w1)[0]
    syn2 = wordnet.synsets(w2)[0]
    
    #print (syn1.name())
    #print (syn2.name())
    result=syn1.wup_similarity(syn2)
    if not result:
        result=0.0
    return result

def check_similarity(word,pos,f1):
    with open(f1, 'r') as reader:
        i=0
        for line in reader.readlines():
            if pos=="PRON":
                lin=line[:len(line)-1].split(':')
                if word==lin[1]:
                    return lin[0]
                else:
                    return "not"
            if i==0:
                i=1
                continue
            lin=line[:len(line)-1].split(':')
            if word==lin[0]:
                return lin[0]
            
            else:
                sim_list=lin[1].split("/")
                if(word in sim_list):
                    return lin[0]
    return "not"

def Select_Level(level):
    if level==1:
        photo="C:/Users/EmyAbobakr/gehad/a"
        f1="C:/Users/EmyAbobakr/gehad/a1.txt"
        f2="C:/Users/EmyAbobakr/gehad/a2.txt"
    elif level==2:
        photo="C:/Users/EmyAbobakr/gehad/b"
        f1="C:/Users/EmyAbobakr/gehad/b1.txt"
        f2="C:/Users/EmyAbobakr/gehad/b2.txt"
    else:
        photo="C:/Users/EmyAbobakr/gehad/c"
        f1="C:/Users/EmyAbobakr/gehad/c1.txt"
        f2="C:/Users/EmyAbobakr/gehad/c2.txt"
    
    return photo,f1,f2

#chack the names of pos
def Check_relation_Errors(relations,ps,data,s_list,s_type,f1):
    i=-1
    wait=[]
    v2=[]
    r=[]
    for words in relations:
        i+=1
        w1=check_similarity(words[0],ps[i][0],f1)
        w2=check_similarity(words[1],ps[i][1],f1)
        if (w1!="not" and w2!="not"):
            rel=data[s_list.index(w1)][s_list.index(w2)].split('=')
            if rel != "0": 
                if ps[i][0]=="V" and ps[i][1]=="V":
                    found=0
                    Rel_Result=data[s_list.index(w1)][s_list.index(w2)].split('=')
                    if (Rel_Result!= "0"):
                        found=1
                        
                    if found==0:
                        return "NOT RELATED"
                    
                elif (ps[i][0]=="V"):
                    #if not question (i)
                    if i>0:
                        #if there is verb before me
                        count=-1
                        found=0
                        for v in v2:
                            count+=1
                            if(v[1]==w1):
                                wait[count]=wait[count]-1
                                Rel_Result=data[s_list.index(v[0])][s_list.index(w2)].split('=')
                                if ((w1==Rel_Result[0] or w1==Rel_Result[0][1:]) and Rel_Result[1]==s_type):
                                    found=1
                                    break
                        
                        if found==0:
                            return "NOT RELATED"
                    
                        if ps[i][1]=="V":
                            v2.append((w1,w2))
                            r.append(rel)
                            if(rel[0][0]=="+"):
                                wait.append(1)
                            else:
                                wait.append(0)   
                        
                    else:
                        return "NOT RELATED"
                    
                    
                elif ps[i][0]!="V" and ps[i][1]=="V":
                    v2.append((w1,w2))
                    r.append(rel)
                    if(rel[0]=="+"):
                        wait.append(1)
                    else:
                        wait.append(0)
                        
            else:
                return "NOT RELATED"
        
        else:
            return "NOT RELATED"
    
    
    for w in wait:
        if w==1:
            return "NOT RELATED"
    
    return "RELATED"

def CHECK(sentence,level):
    e,input_list,ps,s_type=Check_Sentence_Errors(sentence)
    photo,f1,f2=Select_Level(level)
    data,s_list=file_data(f2)
    er=Check_relation_Errors(input_list,ps,data,s_list,s_type,f1)
    if (e[0]=="E" or er[0]=="N"):
        print("---------------",e)
        print("---------------",er)
        return "NOT CORRECT"
    else:
        return "CORRECT"

```
